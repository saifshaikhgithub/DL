```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow import keras
from keras.models import Model, Sequential
from keras.layers import MaxPooling2D, Dropout, Dense, Flatten
```

```python
df = pd.read_csv('ecg.csv', header=None)
```

```python
x = df.drop(140, axis=1)
```

```python
y = df[140]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,  test_size=0.15, stratify=y)
```

```python
x_train
```

```python
train_index = y_train[y_train==1].index
```

```python
train_index
```

```python
x_train.index
```

```python
train_data = x_train.loc[train_index]
```

```python
train_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler(feature_range=(0,1))
```

```python
min_max_scaler
```

```python
x_train_scaled = min_max_scaler.fit_transform(train_data.copy())
```

```python
x_test_scaled = min_max_scaler.transform(x_test.copy())
```

```python
x_test_scaled
```

```python
x_train_scaled
```

```python
class Autoencoder(Model):
    def __init__(self, output_units, code_size=8):
        super().__init__()

        self.encoder = Sequential([
            Dense(64, activation='relu'),
            Dropout(0.2),
            Dense(32, activation='relu'),
            Dropout(0.2),
            Dense(16, activation='relu'),
            Dropout(0.2),
            Dense(code_size, activation='relu')
        ])

        self.decoder = Sequential([
            Dense(16, activation='relu'),
            Dropout(0.2),
            Dense(32, activation='relu'),
            Dropout(0.2),
            Dense(64, activation='relu'),
            Dropout(0.2),
            Dense(output_units, activation='sigmoid')
        ])

    def call(self, input_units):
        encode = self.encoder(input_units)
        decode = self.decoder(encode)
        return decode
```

```python
x_train_scaled.shape
```

```python
model = Autoencoder(output_units=x_train_scaled.shape[1])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='msle', metrics=['mse'])
```

```python
print(x_train_scaled.shape)  # Ensure matches the model's expected input
print(x_test_scaled.shape)
```

```python
history = model.fit(x_train_scaled, x_train_scaled, batch_size=512, validation_data=(x_test_scaled, x_test_scaled), epochs=30)
```

```python
pred1 = model.predict(x_train_scaled)
```

```python
pred1_error = keras.losses.msle(pred1, x_train_scaled)
```

```python
pred1_error
```

```python
threshold = np.mean(pred1_error.numpy()) + np.std(pred1_error.numpy()) 
```

```python
threshold
```

```python
pred2 = model.predict(x_test_scaled)
```

```python
pred2_error = keras.losses.msle(pred2, x_test_scaled)
```

```python
ana = pd.Series(pred2_error) > threshold
```

```python
pre = np.where(ana, 0.0, 1.0)
```

```python
ana
```
